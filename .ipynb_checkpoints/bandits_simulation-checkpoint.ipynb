# 1. Use survival as reward

```python
patients = pd.read_csv('data/patient.csv', usecols=['patientunitstayid','age','gender'])

apache = pd.read_csv('data/apachePatientResult.csv', usecols=['patientunitstayid','actualhospitalmortality','actualiculos']
)

# Map hospital mortality → survival reward (1=survived, 0=died)
apache['reward_surv'] = apache['actualhospitalmortality'].map({0:1, 1:0})

apache['icu_los_days'] = apache['actualiculos'] / 24.0

df_main = patients.merge(apache[['patientunitstayid','reward_surv','icu_los_days']], on='patientunitstayid', how='inner')
```

```python
v1 = pd.read_csv('data/vitalPeriodic.csv', usecols=['patientunitstayid','observationoffset'])
v2 = pd.read_csv('data/vitalAperiodic.csv', usecols=['patientunitstayid','observationoffset'])
vitals = pd.concat([v1, v2], ignore_index=True)
icu_times = (
    vitals
    .groupby('patientunitstayid')['observationoffset']
    .agg(icu_start_h='min', icu_end_h='max')
    .reset_index()
)

patients = pd.read_csv('data/patient.csv', usecols=['patientunitstayid','age','gender'])

apache = pd.read_csv(
    'data/apachePatientResult.csv',
    usecols=['patientunitstayid',
             'actualhospitalmortality',
             'actualicumortality',
             'actualiculos']
)

mort = apache['actualhospitalmortality'].combine_first(apache['actualicumortality'])

mort = mort.replace({'ALIVE': 0, 'EXPIRED': 1})

mort = pd.to_numeric(mort, errors='coerce')
apache['reward_surv'] = 1.0 - mort

apache['icu_los_days'] = apache['actualiculos'] / 24.0

print("APR reward_surv value_counts:")
print(apache['reward_surv'].value_counts(dropna=False).head())

df_main = (
    icu_times
    .merge(patients, on='patientunitstayid', how='inner')
    .merge(apache[['patientunitstayid','reward_surv','icu_los_days']],
           on='patientunitstayid', how='inner')
)

df_main = df_main[
    (df_main['icu_los_days'] > 0) & (df_main['icu_los_days'] < 60)
].sort_values('icu_start_h').reset_index(drop=True)

print("df_main shape:", df_main.shape)
print("→ reward_surv non‐NaN:", df_main['reward_surv'].notna().sum(), "/", len(df_main))
```

```python
print("APR reward_surv value_counts:")
print(apache['reward_surv'].value_counts(dropna=False).head())
```

```python
df_main = (
    icu_times
    .merge(patients, on='patientunitstayid')
    .merge(apache[['patientunitstayid','reward_surv','icu_los_days']],
           on='patientunitstayid')
)
df_main = df_main[(df_main['icu_los_days']>0)&(df_main['icu_los_days']<60)]
df_main.sort_values('icu_start_h', inplace=True)
stream_order = df_main['patientunitstayid'].tolist()
```

```python
def agg_feature(df, tcol, vcol, hours=6):
    recs = []
    for pid, grp in df.groupby('patientunitstayid'):
        if pid not in df_main['patientunitstayid'].values: continue
        start = df_main.loc[df_main['patientunitstayid']==pid, 'icu_start_h'].iat[0]
        mask  = (grp[tcol]>=start) & (grp[tcol]<=start+60*hours)
        vals  = pd.to_numeric(grp.loc[mask, vcol], errors='coerce').dropna()
        if vals.empty: continue
        recs.append({
            'patientunitstayid': pid,
            f'{vcol}_mean': vals.mean(),
            f'{vcol}_std' : vals.std(ddof=0)
        })
    return pd.DataFrame(recs)

vital_df = pd.read_csv('data/vitalPeriodic.csv',
                       usecols=['patientunitstayid','observationoffset','temperature','heartrate','respiration'])
lab_df   = pd.read_csv('data/lab.csv',
                       usecols=['patientunitstayid','labresultoffset','labresult'])

vitals_temp = agg_feature(vital_df, 'observationoffset', 'temperature')
vitals_hr   = agg_feature(vital_df, 'observationoffset', 'heartrate')
vitals_resp = agg_feature(vital_df, 'observationoffset', 'respiration')
labs_agg    = agg_feature(lab_df,   'labresultoffset', 'labresult')
```

```python
contexts = df_main[['patientunitstayid','age','gender']].copy()

contexts['age'] = pd.to_numeric(contexts['age'], errors='coerce')

for feat in (vitals_temp, vitals_hr, vitals_resp, labs_agg):
    contexts = contexts.merge(feat, on='patientunitstayid', how='left')

for col in contexts.columns:
    if col.endswith('_mean') or col.endswith('_std'):
        contexts[col] = pd.to_numeric(contexts[col], errors='coerce')

numeric_cols = contexts.select_dtypes(include=np.number).columns
contexts[numeric_cols] = contexts[numeric_cols].fillna(contexts[numeric_cols].mean())

contexts['gender'] = contexts['gender'].map({'M':1,'F':0})

contexts = contexts.drop_duplicates('patientunitstayid').set_index('patientunitstayid')
```

```python
def map_drug(name):
    if not isinstance(name, str):
        return None
    n = name.lower()
    if 'sodium chloride' in n or '0.9%' in n:
        return 'fluid'
    if any(x in n for x in ['acetaminophen','tylenol','morphine']):
        return 'analgesic'
    if 'ondansetron' in n or 'zofran' in n:
        return 'antiemetic'
    return None

meds = pd.read_csv(
    'data/medication.csv',
    usecols=['patientunitstayid','drugname','drugstartoffset']
).rename(columns={'drugstartoffset':'startoffset'})

first_med = (
    meds.sort_values(['patientunitstayid','startoffset'])
        .groupby('patientunitstayid')
        .first()
        .reset_index()
)
first_med['arm'] = first_med['drugname'].apply(map_drug)
first_med = first_med.dropna(subset=['arm'])

data = (
    contexts
    .merge(first_med[['patientunitstayid','arm']], on='patientunitstayid', how='inner')
    .merge(df_main[['patientunitstayid','reward_surv','icu_los_days']], on='patientunitstayid', how='inner')
)

data = data.set_index('patientunitstayid')

print("After merge, data shape:", data.shape)
print("Survival rewards non‐missing:", data['reward_surv'].notna().sum(), "/", len(data))
print("Arm distribution:\n", data['arm'].value_counts())
```

```python
X = contexts.loc[data.index]
X = X.select_dtypes(include=np.number)
y = data['arm']

print("X shape:", X.shape, "— should be fully numeric")

pipe = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler(),
    LogisticRegression(max_iter=5000, solver='saga', class_weight='balanced')
)
pipe.fit(X, y)

imp   = pipe.named_steps['simpleimputer']
model = pipe.named_steps['logisticregression']
X_imp = imp.transform(X)
probs = model.predict_proba(X_imp)

prop_df = pd.DataFrame(probs, columns=model.classes_, index=X.index)
print("prop_df shape:", prop_df.shape)

from sklearn.ensemble import RandomForestRegressor

rf_surv = RandomForestRegressor(
    n_estimators=200,
    max_depth=None,
    min_samples_leaf=5,
    n_jobs=-1,
    oob_score=True,
    random_state=0
)
rf_surv.fit(X_imp, data['reward_surv'])

print(f"Random-Forest reward model: "
      f"{len(rf_surv.estimators_)} trees  |  "
      f"OOB-R² = {getattr(rf_surv,'oob_score_',np.nan):.3f}")
```

```python
class UCB1:
    def __init__(s, arms):
        s.counts={a:0 for a in arms}; s.values={a:0.0 for a in arms}; s.total=0
    def select(s):
        for a in s.counts:
            if s.counts[a]==0: return a
        scores={a: s.values[a]+math.sqrt(2*math.log(s.total)/s.counts[a])
                for a in s.counts}
        return max(scores, key=scores.get)
    def update(s,a,r):
        s.total+=1; s.counts[a]+=1
        n=s.counts[a]; s.values[a]=((n-1)*s.values[a]+r)/n

class LinUCB:
    def __init__(s, arms, d, α=1.0):
        s.alpha=α; s.arms=arms
        s.A={a:np.eye(d) for a in arms}; s.b={a:np.zeros(d) for a in arms}
    def select(s,x):
        scores={}
        for a in s.arms:
            Ainv=np.linalg.inv(s.A[a])
            θ=Ainv.dot(s.b[a])
            scores[a]=θ.dot(x)+s.alpha*np.sqrt(x.dot(Ainv).dot(x))
        return max(scores, key=scores.get)
    def update(s,a,x,r):
        s.A[a]+=np.outer(x,x); s.b[a]+=r*x

class ThompsonSampling:
    def __init__(s, arms, d, v=1.0):
        s.v=v; s.arms=arms
        s.A={a:np.eye(d) for a in arms}; s.b={a:np.zeros(d) for a in arms}
    def select(s,x):
        samples={}
        for a in s.arms:
            cov=np.linalg.inv(s.A[a])
            mu =cov.dot(s.b[a])
            w  =np.random.multivariate_normal(mu, s.v*cov)
            samples[a]=w.dot(x)
        return max(samples, key=samples.get)
    def update(s,a,x,r):
        s.A[a]+=np.outer(x,x); s.b[a]+=r*x


valid_pids = [pid for pid in stream_order if pid in data.index]
arms       = list(data['arm'].unique())
d          = X_imp.shape[1]

bandits = {
    'UCB1'     : UCB1(arms),
    'LinUCB'   : LinUCB(arms, d),
    'Thompson' : ThompsonSampling(arms, d)
}

history = []
for idx, pid in enumerate(valid_pids):
    row = data.loc[pid]
    if isinstance(row, pd.DataFrame): row = row.iloc[0]
    logged = row['arm']
    r_s = float(row['reward_surv'])
    # r_s = float(row['r_c']) 
    p0 = float(prop_df.iloc[idx][logged])
    x = X_imp[idx]

    for name, b in bandits.items():
        chosen = b.select() if name=='UCB1' else b.select(x)

        if name == 'UCB1':
            b.update(logged, r_s)
        else:
            b.update(logged, x, r_s)

        history.append({
            'policy': name,
            'pid'   : pid,
            'chosen': chosen,
            'logged': logged,
            'prop'  : p0,
            'r_s'   : r_s
        })

hist = pd.DataFrame(history)
print("Simulation records:", hist.shape)
for name in bandits:
    sub = hist[hist['policy']==name]
    matches = (sub['chosen']==sub['logged']).sum()
    print(f"{name}: {matches}/{len(sub)} matches")
```

```python
import numpy as np
import matplotlib.pyplot as plt

EPS = 0.1
r_bar = data['reward_surv'].mean()

def _logged_pairs(policy):
    sub = hist[hist['policy'] == policy]
    mask = sub['chosen'] == sub['logged']
    r = sub.loc[mask, 'r_s'].astype(float).values
    p = sub.loc[mask, 'prop'].clip(lower=EPS).astype(float).values
    return r, p

def _logged_pairs2(policy):
    sub = hist[hist['policy'] == policy]
    mask = sub['chosen'] == sub['logged']
    r = sub.loc[mask, 'r_s'].astype(float).values
    p = sub.loc[mask, 'prop'].clip(lower=EPS).astype(float).values
    return r, p

def ips(policy):
    r, p = _logged_pairs(policy)
    return np.mean(r / p) if len(r) else np.nan

def ips_stabil(policy):
    r, p = _logged_pairs(policy)
    w = r / p
    w /= w.mean()
    return np.mean(w)

def dr(policy):
    """Doubly‐Robust with baseline r̄ and clipped p."""
    r, p = _logged_pairs(policy)
    rhat = r_bar
    correction = (r - rhat) / p
    return float(rhat + correction.mean()) if len(r) else np.nan

print("{:<10s} {:>12s} {:>12s} {:>12s}".format("Policy", "IPS", "IPS-stab", "DR"))
for pol in bandits:
    print(
        "{:<10s} {:12.4f} {:12.4f} {:12.4f}".format(
            pol, ips(pol), ips_stabil(pol), dr(pol)
        )
    )

valid_pids = [pid for pid in stream_order if pid in data.index]
cum = {p: [] for p in bandits}
tot = {p: 0.0 for p in bandits}

for pid in valid_pids:
    for p in bandits:
        rec = hist[(hist.policy == p) & (hist.pid == pid)].iloc[0]
        if rec.chosen == rec.logged:
            tot[p] += rec.r_s
    for p in bandits:
        cum[p].append(tot[p])

plt.figure()
for p, v in cum.items():
    plt.plot(v, label=p)
plt.legend()
plt.title("Cumulative Reward")
plt.xlabel("Patient #")
plt.ylabel("Reward")
plt.show()

plt.figure()
M = np.vstack([cum[p] for p in bandits])
best = M.max(axis=0)
for i, p in enumerate(bandits):
    plt.plot(best - M[i], label=p)
plt.legend()
plt.title("Cumulative Regret")
plt.xlabel("Patient #")
plt.ylabel("Regret")
plt.show()
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cell T-2  Bandit simulation with LoS reward
# (identical structure to the survival run – only two lines change)

valid_pids = [pid for pid in stream_order if pid in data.index]
d          = X_imp.shape[1]

bandits_t = {
    'UCB1'     : UCB1(arms),
    'LinUCB'   : LinUCB(arms, d),
    'Thompson' : ThompsonSampling(arms, d)
}

hist_t = []
for pid in valid_pids:
    row    = data.loc[pid]
    logged = row['arm']
    r_t    = float(row['reward_los'])                 # ❶  LoS reward
    p0     = float(prop_df.at[pid, logged])           # ❷  use .at with pid key
    x      = X_imp[data.index.get_loc(pid)]           # ❸  grab the same row in X_imp

    for name, b in bandits_t.items():
        chosen = b.select() if name == 'UCB1' else b.select(x)

        # always update on the logged action
        if name == 'UCB1':
            b.update(logged, r_t)
        else:
            b.update(logged, x, r_t)

        hist_t.append({
            'policy': name,
            'pid'   : pid,
            'chosen': chosen,
            'logged': logged,
            'prop'  : p0,
            'r_t'   : r_t
        })

hist_t = pd.DataFrame(hist_t)
print("LoS simulation rows:", hist_t.shape[0])
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cell T-3 IPS · Stabilised IPS · DR  (LoS reward)
EPS = 0.1                       # same clip as before
μ_t  = data['reward_los'].mean() # baseline for DR


def _pairs_t(pol):
    sub  = hist_t[hist_t.policy == pol]
    msk  = sub.chosen == sub.logged
    r    = sub.loc[msk, 'r_t'].values
    p    = sub.loc[msk, 'prop'].clip(lower=EPS).values
    return r, p

def ips_t(pol):
    r, p = _pairs_t(pol)
    return np.mean(r / p) if len(r) else np.nan

def ips_stab_t(pol):
    r, p = _pairs_t(pol)
    w    = r / p
    return np.mean(w / w.mean()) if len(w) else np.nan

def dr_t(pol):
    r, p = _pairs_t(pol)
    return μ_t + np.mean((r - μ_t) / p) if len(r) else np.nan

print("{:<10} {:>10} {:>10} {:>10}".format("Policy","IPS","SN-IPS","DR"))
for pol in bandits_t:
    print("{:<10} {:10.4f} {:10.4f} {:10.4f}"
          .format(pol, ips_t(pol), ips_stab_t(pol), dr_t(pol)))
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# Cell T-4 Cumulative reward / regret plots (LoS reward)

cum_t   = {p: [] for p in bandits_t}
tot_t   = {p: 0.0 for p in bandits_t}

for pid in valid_pids:
    for p in bandits_t:
        rec = hist_t[(hist_t.policy == p) & (hist_t.pid == pid)].iloc[0]
        if rec.chosen == rec.logged:
            tot_t[p] += rec.r_t
    for p in bandits_t:
        cum_t[p].append(tot_t[p])

# Reward
plt.figure()
for p, vals in cum_t.items(): plt.plot(vals, label=p)
plt.legend(); plt.xlabel("Patient #"); plt.ylabel("Σ reward_los (higher=better)")
plt.title("Cumulative Reward (LoS)"); plt.show()

# Regret
plt.figure()
M = np.vstack([cum_t[p] for p in bandits_t])
best = M.max(axis=0)
for i, p in enumerate(bandits_t):
    plt.plot(best - M[i], label=p)
plt.legend(); plt.xlabel("Patient #"); plt.ylabel("Regret")
plt.title("Cumulative Regret (LoS)"); plt.show()
```
